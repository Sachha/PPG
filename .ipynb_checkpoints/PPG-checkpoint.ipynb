```python
for k,v in quest.items():
    df[k] = v
```

```python
def splitdata(data):
    splited = np.split(data, 36848)
    temp =  []
    for v in splited:
        temp.append(np.mean(v))
    return temp
#4HZ mesure
df['signal.wrist.EDA'] = data['signal']['wrist']['EDA']
df['signal.wrist.TEMP'] = data['signal']['wrist']['TEMP']
#other mesure
df['signal.wrist.BVP'] = splitdata(data['signal']['wrist']['BVP'])
df['signal.wrist.ACC'] = splitdata(data['signal']['wrist']['ACC'])
df['signal.chest.ACC'] = splitdata(data['signal']['chest']['ACC'])
df['signal.chest.ECG'] = splitdata(data['signal']['chest']['ECG'])
df['signal.chest.Resp'] = splitdata(data['signal']['chest']['Resp'])
```
